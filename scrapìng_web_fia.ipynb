```python
from bs4 import BeautifulSoup
import urllib.request
url = 'https://fiaresultsandstatistics.motorsportstats.com/series/world-endurance-championship'
pagina = urllib.request.urlopen(url)
soup = BeautifulSoup(pagina,'html.parser')  # o usar lxml, html.parser
print(soup.prettify())
```

```python

for h2 in soup.find_all('div',{'class','Vv8Fg'}):
  cabecera = h2.find('h2')

  print(cabecera.text)
  for i in h2.find_all('table'): 
    cabecera = i.find(('tr',{'class','_2Q90P'}))
    # print(cabecera.prettify())
    for j in i.find_all('tr',{'class','_3AoAU'}):
      for k in j.find_all('td',{'class','_2sWDi'}):
        dato = k.find('a')
        if dato != None:
          href = dato['href']
          if href == None:
            href = '---'
          description = dato.text
          print(f'{href} - {description}')
        else:
          dato = k.find('div')
          if dato != None:
            descripcion = dato.text
            print(descripcion)
```
